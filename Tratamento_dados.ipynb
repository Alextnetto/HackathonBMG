```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel("economatica.xlsx", header=3)
TAXA_SELIC = 3.15
```

```python
#Filtro
df.replace({'-': np.nan}, inplace=True)
df.dropna(inplace=True)
df = df[df.Data > '2019-05-30'][['Data', 'Abertura']].reset_index(drop=True)
df.rename(columns={'Abertura': 'Americanas'}, inplace=True)
```

```python
#Tratamento dos dados

df['Poupanca'] = pd.Series([0 for x in range(len(df.index))]).astype(float)
df.Data = df.Data.apply(lambda date: date.strftime('%d/%m/%Y'))
df.Americanas = df.Americanas.astype(float).apply(lambda valor: round(valor, 2))
df
```

```python
RENDIMENTO_POUPANCA_DIA = (TAXA_SELIC / 25000) + 1

valor = 16.20
for i in range(df.shape[0]):
    df.Poupanca[i] = round(valor, 2)
    valor = valor * RENDIMENTO_POUPANCA_DIA

df
```

```python
df.to_json('data.json', orient='records')
```
